```python
import pandas as pd
import requests
```

```python
url='https://api.covid19api.com/summary'
```

```python
r=requests.get(url)
```

```python
data=r.json()
```

```python
data
```

```python
df=pd.DataFrame(data["Countries"])
df
```

```python
i = df["TotalDeaths"].idxmax()
print(i)
x = df["Country"][i]
x
```

```python
k = df["TotalRecovered"].idxmin()
print(k)
f = df["Country"][k]
f
```

```python
total_death_cases=df["TotalDeaths"].sum(axis=0)
total_death_cases
```

```python
total_recovered_cases=df["TotalRecovered"].sum(axis=0)
total_recovered_cases
```

```python
Cmaximumdeaths=df["TotalDeaths"].max(axis=0)
Cmaximumdeaths
```

```python
Cmaxrecovered=df["TotalRecovered"].max(axis=0)
Cmaxrecovered
```

```python
data={"Total Death Cases":total_death_cases,"Total Recovered Cases":total_recovered_cases,"CountrymaxDeath":x,"Country Max Death Cases":Cmaximumdeaths,"Countrymaxrecovered":f,"Country Max Recovered Cases":Cmaxrecovered}
df=pd.DataFrame(data,index=[1])
df
```
